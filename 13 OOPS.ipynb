```python
"""Q5. What is inheritance? Give an example for each type of inheritance."""
#inheritance is pillar of OOPS used to inherit functions from parent class to child class

class A():
    def __init__(self):
        pass
    def  abc(self):
        return "100"

class B(A):
    def __init__(self):
        pass
    def xyz(self):
        return "200"

cls_test=B()
cls_test.abc()


class C():
    def __init__(self):
        pass
    def qwert(self):
        return "qwerty"

class D(A,C):
    def __init__(self):
        pass
    def qwert1(self):
        return "qwerty1"
    
class_test1=D()
class_test1.abc()
```
