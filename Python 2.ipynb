```python
"""Q1. write a program to accept percentage from the user and display the grade according to the following criteria:
   
    Marks	           Grade
    >90 	             A
    >80 and <=90	     B
    >=60 and <=80	     C
    below 60	         D
   
   """

def GradeChecking(percentage):
    if(percentage<=100 and percentage>=0):
        if(percentage>90):
            return "A"
        elif (percentage>80 and percentage<=90):
            return "B"
        elif (percentage>=60 and percentage<=80):
            return "C"
        else:
            return "D"
    else:
        return "Enter valid percentage between 0 to 100"
print(GradeChecking(81))
    
```

```python
"""Q2. Write a program to accept the cost price of a bike and display the road tax to be paid according to the following criteria

 TAX	       COST Price (in Rs)
 15%	          >100000
 10%	          >50000 and <=100000
  5%	          <=50000

"""

def RoadTax(COST):
    if(COST>0):
        if(COST>100000):
            return "15%"
        elif (COST>50000 and COST<=100000):
            return "10%"
        else:
            return "5%"
    else:
        return "Enter valid COST which is greater than 0"
    
RoadTax(200000)
```

```python
"""Q3. Accept any city from the user and display monuments of that city.

City	Monument
Delhi	Red Fort
Agra	Taj Mahal
Jaipur	Jal Mahal

"""

def monuments(city):
    if(str(city).lower()=="delhi"):# I'm converting the input to lower case bcz sometimes use may enter data in lower case
        return "Red Fort"
    elif(str(city).lower()=="agra"):
        return "Taj Mahal"
    elif(str(city).lower()=="jaipur"):
        return "Jal Mahal"
    else:
        return "Enter tvalid city to get monuments"
    
print(monuments("Delhi"))
```

```python
"""Q4. Check how many times a given number can be devided by 3 before it is less then or equal to 10?"""


def check(n):
    i=0
    while True:
        n=n/3
        i+=1
        if(n<=10):
            break
    return i
print(check(1090))
```

```python
"""Q6. Use nested while loop to print 3 different patterns?"""

i = 1
while i <= 5:
    j = 1
    while j <= i:
        print(j, end=' ')
        j += 1
    print()
    i += 1

print("\n\n\n")
    
i = 5
while i >= 1:
    j = 1
    while j <= i:
        print('*', end=' ')
        j += 1
    print()
    i -= 1

print("\n\n\n")
    
i = 1
num = 1
while i <= 4:
    j = 1
    while j <= i:
        print(num, end=' ')
        j += 1
        num += 1
    print()
    i += 1

print("\n\n\n")
```
