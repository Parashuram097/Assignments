```python
"""Q1. Which keyword is used to create a function? Create a function to return a list of odd numbers in the
range of 1 to 25."""

# def keyword is used to define function in Python

def odd_numbers():
    l1=[]
    for i in range(1,26,1):
        if(i%2!=0):
            l1.append(i)
    return l1
odd_numbers()
```

```python
"""Q2. Why *args and **kwargs is used in some functions? Create a function each for *args and **kwargs
to demonstrate their use."""

# *args and **kwargs are used when we don't know the exact length of the parameters are going to be passed.
# *args is tuple
# *kwargs is dictionary.

def args_def(*args):
    for i in args:
        print(i)
        
def kwargs_def(**kwargs):
    for i,j in kwargs.items:
        print(i,j)
```

```python
"""Q4. What is a generator function in python? Why yield keyword is used? Give an example of a generator
function."""

# Generator function used to create function that acts like iterator
# yield keyword is used for returning generator object.

def gerator_assignment():
    n = 0
    while True:
        yield n
        n += 2
gerator_assignment()
```

```python
"""Q5. Create a generator function for prime numbers less than 1000. Use the next() method to print the
first 20 prime numbers."""

def prime_generator():
    prime = [True] * 1000
    for i in range(2, int(1000**0.5) + 1):
        if prime[i]:
            for j in range(i**2, 1000, i):
                prime[j] = False
    for i in range(2, 1000):
        if prime[i]:
            yield i

gen = prime_generator()
for i in range(20):
    print(next(gen))
```

```python
"""Q6. Write a python program to print the first 10 Fibonacci numbers using a while loop."""

a, b = 0, 1
count = 0

while count < 10:
    print(a)
    c = a + b
    a = b
    b = c
    count += 1
```

```python
"""Q8. Write a python program to check whether a given number is Palindrome or not using a while loop."""

n = int(input("Enter a number: "))
temp = n
reverse = 0

while temp > 0:
    digit = temp % 10
    reverse = (reverse * 10) + digit
    temp //= 10

if n == reverse:
    print(n, "is a palindrome")
else:
    print(n, "is not a palindrome")
```
